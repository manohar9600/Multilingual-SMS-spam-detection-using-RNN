# Text Preprocessing

```python
import pandas as pd
import numpy as np
from collections import Counter
from keras.preprocessing.text import text_to_word_sequence
```

```python
data1 = pd.read_csv('Resources/spam.csv', encoding = 'latin-1')
list = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']
data1 = data1.drop(list, axis=1)

data2 = pd.readdata2 = pd.ExcelFile('Resources/revisedindiandataset.xls')
data2 = data2.parse(0)
data2 = data2.drop(['code'], axis=1)

data = pd.concat([data1, data2])
print(data.head())
```

```python
class userfunc:

    def fit_texts(self, data, num_words):
        temp = []
        for row in data.itertuples():
            rev = str(row[2])
            rev = text_to_word_sequence(rev)
            temp.append(rev)
        self.num_words = num_words
        te2 = []
        for i in range(len(temp)):
            for j in range(len(temp[i])):
                te2.append(temp[i][j])
        q = Counter(te2)
        le = q.most_common(num_words - 1)
        temp = le[0:num_words - 1]
        self.words = []
        for item in temp:
            self.words.append(item[0])
        # print(self.words)
        
    def tagging(self, sentence):
        words = text_to_word_sequence(sentence)
        for i in range(len(words)):
            try:
                temp = int(words[i])
                if len(words[i]) == 10 and int(temp / 1000000000) >= 7:
                    words[i] = '<ph>'
                    continue
                if len(words[i]) == 12 and int(temp / 100000000000) == 9 and int(temp / 10000000000) == 1 and int(temp / 1000000000) >= 7:
                    words[i] = '<ph>'
                    continue
                if len(words[i]) == 11 and int(temp / 10000000000) == 0 and int(temp / 1000000000) >= 7:
                    words[i] = '<ph>'
                    continue
            except Exception:
                pass
            if words[i] not in self.words:
                words[i] = '<unk>'
                continue
        return words
```

```python
print(data)
```

```python
func = userfunc()
func.fit_texts(data, 2500)
sentences = []
pro_data = []
for row in data.itertuples():
    rev = str(row[2])
    words = func.tagging(rev)
    
    if(len(words) == 0):
        pro_data.append(words)
        continue
    
    if(len(words) == 1):
        te = ['</s>', words[0], '</e>']
        sentences.append(te)
        pro_data.append(words)
        continue
    
    te = ['</s>', words[0], words[1]]
    sentences.append(te)
    
    for i in range(1,len(words) - 1):
        te = [words[i-1], words[i], words[i+1]]
        sentences.append(te)
        
    te = [words[len(words) - 2], words[len(words) - 1], '</e>']
    sentences.append(te)   
    
    pro_data.append(words)
```

```python
print(pro_data)
```

# converting sentences to vectors

```python
size = 150
from gensim.models import Word2Vec
model = Word2Vec(sentences, min_count=1, size = size)
```

```python
model['papam']
```

```python
data_vec = []
msg_limit = 200
for i in range(len(pro_data)):
    empty = [0 for i in range(size)]
    temp = [[0 for i in range(size)] for j in range(msg_limit)]
    for j in range(len(pro_data[i])-1, -1, -1):
        temp[199 - (len(pro_data[i])-1-j)] = model[pro_data[i][j]]
    data_vec.append(temp)
data_vec = np.array(data_vec)
```

```python
print(data_vec.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    data_vec, data['label'], test_size=0.30, random_state=0)

Y_train = pd.get_dummies(Y_train).values
```

# classifier model

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, Dropout

classifier = Sequential()
classifier.add(LSTM(200, input_shape =(200, size)))
classifier.add(Dropout(0.2))
classifier.add(Dense(2, activation='softmax'))
classifier.compile(
    loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(classifier.summary())
```

```python
classifier.fit(X_train, Y_train, epochs = 15, batch_size = 50)
```

```python
Y_pr = classifier.predict(X_test)

Y_pred = []

for i in range(len(Y_pr)):
    if Y_pr[i][0] >= Y_pr[i][1]:
        Y_pred.append('ham')
    else:
        Y_pred.append('spam')
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)  
print(cm)
```
